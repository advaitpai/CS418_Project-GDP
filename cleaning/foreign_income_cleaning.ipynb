# Net foreign factor income Dataset Cleaning

```python
import pandas as pd
import numpy as np
```

```python
fi = pd.read_csv('../data/uncleaned/foreign_income.csv')
fi.head()
```

```python
fi.shape
```

**This Dataset has about 266 countries records for 62 countries and 3 other features**

```python
fi.nunique()
```

**The Indicator code and name are constant variables, so the data is only dependent on Country and the year of value.**

**Hence, we drop these features.**

```python
fi = fi.drop(columns = ['Country Name','Indicator Name','Indicator Code'])
fi
```

**Now the dataset has countries and years, which we will unpivot to make it fit for feeding it to our model.**

**So, our dataset size before unpivoting and cleaning is 266*62 = 16,492 records**

```python
fi = fi.rename(columns = {"Country Code":"LOCATION"})
```

```python
fi = fi.melt(id_vars=["LOCATION"], 
        var_name="TIME", 
        value_name="Value")
fi
```

```python
fi = fi.dropna()
fi
```

**After dropping the data boils down to 9063 records**

```python
fi.to_csv('../data/temp/foreign_income_cleaned.csv', mode='wb',index=False)
```
