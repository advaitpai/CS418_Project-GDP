```python
group = lb[(lb['frequency']=='A')]
print(group.shape)
```

```python
group
```

```python
min_year = 2001
# int(group['time'].min())
max_year = int(group['time'].max())
sorted_cont = group['location'].unique()
sorted_cont.sort()
for cont in sorted_cont:
    years = []
    for y in range(min_year,max_year):
        if (str(y) not in group[group['location']==cont]['time'].astype("string").values):
            years.append(y)
    if len(years)!=0:
        print(cont, years)
```
