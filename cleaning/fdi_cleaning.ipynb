```python
fdi_in = fdi_in.drop(columns = ['Country Name','Indicator Name','Indicator Code','1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','Unnamed: 66'])
fdi_in
```

```python
fdi_in = fdi_in.rename(columns = {"Country Code":"LOCATION"})
```

## FDI Outflow

```python
fdi_out = pd.read_csv('../data/uncleaned/FDI_outflow.csv')
fdi_out.head()
```

```python
fdi_out.shape
```

```python
fdi_out.nunique()
```

```python
fdi_out = fdi_out.drop(columns = ['Country Name','Indicator Name','Indicator Code','1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','Unnamed: 66'])
fdi_out
```

```python
fdi_out = fdi_out.rename(columns = {"Country Code":"LOCATION"})
```

```python
fdi_out = fdi_out.melt(id_vars=["LOCATION"], 
        var_name="TIME", 
        value_name="FDI_OUT")
fdi_out
```

```python
fdi = pd.merge(fdi_in,fdi_out)
fdi['Value'] = fdi['FDI_IN'] - fdi['FDI_OUT']
fdi
```

```python
fdi=fdi.drop(columns=['FDI_IN','FDI_OUT'])
fdi
```
